```python
import numpy as np
```

```python
def predict(arg_X, arg_W, arg_b):
    Ŷ = arg_X.dot(arg_W) + arg_b
    return Ŷ
```

```python
def error(arg_Y, arg_Ŷ):
    Error = arg_Ŷ - arg_Y
    return Error
```

```python
def mse_loss(arg_Y, arg_Ŷ):
    Loss = error(arg_Y, arg_Ŷ) **2
    return Loss
```

```python
def mse_cost(arg_Y, arg_Ŷ):
    m = len(arg_Y)
    cost = sum(mse_loss(arg_Y, arg_Ŷ))
    cost /= m
    return (cost)
```

```python
def fit(arg_X, arg_Y, no_of_iterations, arg_initial_b, arg_initial_w, arg_learning_rate):
    no_of_features = arg_X.shape[1]
    b = 0
    W = np.full(no_of_features, arg_initial_w, dtype=np.float64)
    
    for iteration in range(no_of_iterations):
        Ŷ = predict(arg_X, W, b)
        b -= compute_gradient_bias(arg_X, arg_Y, Ŷ) * arg_learning_rate
        W -= compute_gradient_weights(arg_X, arg_Y, Ŷ) * arg_learning_rate
        print(f"{b}, {W}, {mse_cost(arg_Y, Ŷ):0.4f}")
        
    return (b, W)
```

```python
def compute_gradient_bias(arg_X, arg_y_i, arg_ŷ_i):
    m = arg_X.shape[1]
    Error = error(arg_y_i, arg_ŷ_i)
    derivative_bias = sum(Error) * 2 / m
    return derivative_bias

def compute_gradient_weights(arg_X, arg_y_i, arg_ŷ_i):
    m = arg_X.shape[1]
    Error = error(arg_y_i, arg_ŷ_i)
    derivative_weight = Error.dot(arg_X) * 2 / m
    return derivative_weight
```

```python
X = np.array([[1,2], [1,2]])
Y = np.array([2,2])
# Why the hell does it overshoot like this crazy
fit(X, Y, no_of_iterations = 25, arg_initial_b = 0, arg_initial_w = 0, arg_learning_rate = 1)
print()
fit(X, Y, no_of_iterations = 25, arg_initial_b = 0, arg_initial_w = 0, arg_learning_rate = .325)
print()
fit(X, Y, no_of_iterations = 25, arg_initial_b = 0, arg_initial_w = 0, arg_learning_rate = .1)
```

```python
X = np.array([[1,2], [2,4], [3,6]])
Y = np.array([2, 8, 18])
b, W = fit(X, Y, no_of_iterations = 25, arg_initial_b = 0, arg_initial_w = 0, arg_learning_rate = .01)

X_test = np.array([[3, 3]])

print(predict(X_test, W, b))
```
